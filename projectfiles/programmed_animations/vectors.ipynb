```python
#Plot initial point
plt.close('all')
plt.scatter(.5,.5)
plt.xlim(0,12)
plt.ylim(0,6)
plt.axis('off')
plt.text(0.25, 0,"Start")
Ncurrent = 0
Ntotal = 15
for i in tqdm(range(Ncurrent, Ntotal)):
    plt.savefig("frame_%05d.png"%i,dpi=300,bbox_inches='tight')
    clear_output(wait=True)
    display(plt.gcf())
```

```python
Ncurrent = Ntotal
Ntotal += 30
for i in tqdm(range(Ncurrent, Ntotal)):
    ts = i
    plt.close('all')
    plt.scatter(.5,.5)
    plt.scatter(.5+4.5*(ts-Ncurrent)/(Ntotal-Ncurrent),.5+4.5*(ts-Ncurrent)/(Ntotal-Ncurrent))
    plt.xlim(0,12)
    plt.ylim(0,6)
    plt.text(0.25, 0,"Start")
    #plt.plot(ts,t/s%2500)
    plt.axis('off')
    #plt.text(10,n/2*0.8, "            t = %s"%i, color='C0')
    #plt.text(10,n/2*0.7, "t_repeat = t %% 2500 = %s"%(i%2500), color='C1')
    plt.savefig("frame_%05d.png"%i,dpi=300,bbox_inches='tight')
    f = plt.gcf()
    clear_output(wait=True)
    display(plt.gcf())
```
